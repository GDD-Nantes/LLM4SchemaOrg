```python
import pandas as pd
from itertools import chain
from collections_extended import setlist
from rdflib import ConjunctiveGraph

import pandas as pd
import mapply

mapply.init(
    n_workers=-1,
    chunk_size=1,
    max_chunks_per_worker=8,
    progressbar=True
)
```

```python
#@title Load Schema.org vocabulary into a graph
SCHEMA_GRAPH = ConjunctiveGraph()
SCHEMA_GRAPH.parse("https://raw.githubusercontent.com/schemaorg/schemaorg/main/data/releases/23.0/schemaorg-all-http.ttl", format="turtle")
```

```python
df = pd.read_parquet("stats.parquet")
df["sample_feature"] = df["document_path"].str.extract(r"(pset_length|count_sum)")
df["stratum"] = df["document_path"].str.extract(r"(stratum_\d+)")
df.set_index(keys="document_path", inplace=True)
df
```

# Statistics about the type

## Distribution of types

```python
#@title Coverage w.r.t schema.org types

qres = SCHEMA_GRAPH.query("""
SELECT ?type ?label
WHERE {
    ?type a rdfs:Class .
    ?type rdfs:label ?label .
}
""")

all_schemaorg_types = set([label.toPython() for _, label in qres ])

def num_types_distinct(row):

    all_classes = list(chain(*row["document_sub_types"]))

    all_classes = (
        set(chain(*row["document_main_types"])) |
        set(chain(*row["document_sub_types"]))
    )

    row["#types"] = (all_classes & all_schemaorg_types)
    return row

def num_types(row):

    all_classes = list(chain(*row["document_sub_types"]))
    row["#types"] = [ c for c in all_classes if c in all_schemaorg_types ]
    return row

```

```python
#@title Coverage w.r.t schema.org properties
qres = SCHEMA_GRAPH.query("""
SELECT ?type ?label
WHERE {
    ?type a rdf:Property .
    ?type rdfs:label ?label .
}
""")

all_schemaorg_properties = set([label.toPython() for _, label in qres ])

def num_props_distinct(row):

    all_props = set(row["document_props"])

    # print(f"Coverage: {len(all_props & all_schemaorg_properties)} / {len(all_schemaorg_properties)}")
    row["#props"] = (all_props & all_schemaorg_properties)
    return row

def num_props(row):
    all_props = set(row["document_props"])
    row["#props"] = [ p for p in all_props if p in all_schemaorg_properties ]
    return row
```

```python
all_props = set(df["document_props"].explode().to_list())
print(f"Coverage: {len(all_props & all_schemaorg_properties)} / {len(all_schemaorg_properties)}")
```

```python
#@title Number of triples

def number_of_triples(row):
    row["#triples"] = len(row["document_props"])
    return row
```

# Final results

```python
list_df = (df
    .mapply(num_props, axis=1)
    .mapply(num_types, axis=1)
    .mapply(number_of_triples, axis=1)
    .drop(["document_id", "document_main_types", "document_sub_types", "document_props"], axis=1)
)
```

```python
def flatten(x):
    return list(chain(*x))

list_df.reset_index().pivot_table(
    values=["#props", "#types", "#triples", "document_size_bytes"],
    index=["model", "metric"],
    aggfunc = {
        "#triples": lambda x: f"{min(x)}/{int(sum(x)/len(x))}/{max(x)}",
        "document_size_bytes": lambda x: f"{min(x)}/{int(sum(x)/len(x))}/{max(x)}",
        "#props": lambda x: len(flatten(x)),
        "#types": lambda x: len(flatten(x)),
    },
    margins=True,
    margins_name="Overview"
)
```

```python
set_df = (df
    .mapply(num_props_distinct, axis=1)
    .mapply(num_types_distinct, axis=1)
    .mapply(number_of_triples, axis=1)
    .drop(["document_id", "document_main_types", "document_sub_types", "document_props"], axis=1)
)
```

```python
def union(x):
    return set().union(*x)

set_df.query("`model` == 'baseline' and `metric` == 'input'").reset_index().pivot_table(
    values=["#props", "#types", "#triples", "document_size_bytes"],
    index=["sample_feature", "stratum"],
    aggfunc = {
        "#triples": lambda x: f"{min(x)}/{int(sum(x)/len(x))}/{max(x)}",
        "document_size_bytes": lambda x: f"{min(x)}/{int(sum(x)/len(x))}/{max(x)}",
        "#props": lambda x: len(union(x)),
        "#types": lambda x: len(union(x)),
    },
    margins=True,
    margins_name="Overview"
)
```
