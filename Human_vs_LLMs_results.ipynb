```python
import pandas as pd
schemaorg_props = pd.read_csv("schemaorg-current-http-properties.csv")
schemaorg_props
```

```python
!pip install llm_cost_estimation
```

```python

from llm_cost_estimation import count_tokens

model= "gpt-4"
```

```python
def estimate_token_count_prop(row):
    tok_str = f"{row['label']} ( expected_type={row['rangeIncludes']} ) "
    tok_domain = f"{row['label']} ( expected_type={row['rangeIncludes']}, domain={row['domainIncludes']} )"
    tok_domain_comment = f"{row['label']} ( expected_type={row['rangeIncludes']}, domain={row['domainIncludes']}, comment={row['comment']} )"
    tok_count_1, tok_cost_1 = count_tokens(tok_str, model)
    tok_count_2, tok_cost_2 = count_tokens(tok_domain, model)
    tok_count_3, tok_cost_3 = count_tokens(tok_domain_comment, model)
    row["tok_count"] = tok_count_1
    row["tok_cost"] = tok_cost_1
    row["tok_str"] = tok_str
    row["tok_count_domain"] = tok_count_2
    row["tok_cost_domain"] = tok_cost_2
    row["tok_count_domain_comment"] = tok_count_3
    row["tok_cost_domain_comment"] = tok_cost_3

    return row

stats = schemaorg_props.fillna(value="").apply(estimate_token_count_prop, axis=1)
```
